### Defining hyperparameter spaces

```python
from hyperopt import fmin, tpe, hp, STATUS_OK, space_eval
reload(h)

# SET REFERENCE CLASS HERE
reference_class = 'nt' # (tnt or nt)

# Initialise dicts
ml_metrics_tnt = {} 
ml_metrics_nt = {}
ml_best_params_tnt = {}
ml_best_params_nt = {}
space_dict = {} # Store the search space for each classifier

# ------------ XGBoost ------------

# Define the hyperparameter space
dataset_name = h.get_dataset_name(8, reference_class)
X_train, y_train, X_val, y_val, _, _, _, _, _, _, _, _, _, _, _ = h.prepare_datasets(filename=f"output/{dataset_name}", test_size=0.1, val_size=0.1, verbose=False)
counts = y_train.value_counts()
scale_pos_weight = counts[0] / counts[1] # Recommended by: https://webcache.googleusercontent.com/search?q=cache:https://towardsdatascience.com/a-guide-to-xgboost-hyperparameters-87980c7f44a9&sca_esv=254eb9c569a53dbc&strip=1&vwsrc=0

# Default recommendations: https://bradleyboehmke.github.io/xgboost_databricks_tuning/tutorial_docs/xgboost_hyperopt.html

xgb_space = {
    'window_size': hp.choice('window_size', range(8, 34, 2)),
    # 'reference_class': hp.choice('reference_class', ['tnt', 'nt']),
    'objective':'binary:logistic',
    'max_depth': hp.choice('max_depth', np.arange(2, 11, dtype=int)),
    'min_child_weight': hp.uniform('min_child_weight', 0.1, 15),
    'learning_rate': hp.loguniform('learning_rate', np.log(0.0001), np.log(1)),
    'subsample': hp.uniform('subsample', 0.5, 1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1),
    'colsample_bylevel': hp.uniform('colsample_bylevel', 0.5, 1),
    'colsample_bynode': hp.uniform('colsample_bynode', 0.5, 1),
    'n_estimators': hp.choice('n_estimators', range(50, 5000)),
    'gamma': hp.choice('gamma', [0, hp.loguniform('gamma_log', np.log(1), np.log(1000))]),
    'reg_lambda': hp.choice('reg_lambda', [0, hp.loguniform('reg_lambda_log', np.log(1), np.log(1000))]),
    'reg_alpha': hp.choice('reg_alpha', [0, hp.loguniform('reg_alpha_log', np.log(1), np.log(1000))]),
    'scale_pos_weight': scale_pos_weight
}

space_dict['xgb'] = xgb_space

# ------------ GLM ------------

# Define the hyperparameter space
glm_space = {
    'window_size': hp.choice('window_size', range(8, 34, 2)),
    'C': hp.loguniform('C', np.log(0.001), np.log(1000)),
    'penalty': hp.choice('penalty', ['l1', 'l2']),
    'solver': hp.choice('solver', ['liblinear', 'saga']), # Only solvers that support both L1 and L2 penalties
    'class_weight' : 'balanced',
    'max_iter': 10000
}

space_dict['glm'] = glm_space

# ------------ Random Forest ------------

# Source: https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74
rf_space = {
    'window_size': hp.choice('window_size', range(8, 34, 2)),
    'n_estimators': hp.choice('n_estimators', range(2, 200)),
    'max_depth': hp.choice('max_depth', np.arange(2, 101, dtype=int)),
    'max_features': hp.choice('max_features', ['log2', 'sqrt', None]),
    'min_samples_split': hp.choice('min_samples_split', np.arange(2, 10, dtype=int)),
    'min_samples_leaf': hp.choice('min_samples_leaf', np.arange(1, 5, dtype=int)),
    'class_weight' : 'balanced'
}

space_dict['rf'] = rf_space

# ------------ SVM ------------

# Define the hyperparameter space
svm_space = hp.choice('model_type', [
    {
        'window_size': hp.choice('window_size_linear', range(8, 34, 2)),
        'C': hp.loguniform('C_linear', np.log(0.01), np.log(10)),  # Lower range for C
        'kernel': 'linear',
        'class_weight' : 'balanced'
    },
    {
        'window_size': hp.choice('window_size_rbf', range(8, 34, 2)),
        'C': hp.loguniform('C_rbf', np.log(0.01), np.log(10)),
        'kernel': 'rbf',
        'gamma': hp.loguniform('gamma_rbf', np.log(0.001), np.log(1)),  # Lower range for gamma
        'class_weight' : 'balanced'
    }
])

space_dict['svm'] = svm_space
```

### Without cross-validation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import average_precision_score
from hyperopt import fmin, tpe, STATUS_OK
reload(h)

def optimise_model(model_name, space, max_evals=100):
    # Map the model name to the corresponding classifier
    model_dict = {
        'xgb': XGBClassifier,
        'glm': LogisticRegression,
        'rf': RandomForestClassifier,
        'svm': SVC
    }
    model = model_dict[model_name]

    # Define the objective function to minimize
    def objective(params):
        # Load the appropriate dataset
        window_size = params.pop('window_size')
        dataset_name = h.get_dataset_name(window_size, reference_class)
        X_train, y_train, X_val, y_val, _, _, _, _, _, _, _, _, _, _, participants = h.prepare_datasets(filename=f"output/{dataset_name}", test_size=0.1, val_size=0.1, verbose=False)

        # Train and evaluate the model
        clf = model(**params, random_state=42)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_val)
        score = average_precision_score(y_val, y_pred)

        # Add the window size back into params for logging
        params['window_size'] = window_size

        return {'loss': -score, 'status': STATUS_OK, 'params': params}

    # Perform the optimisation
    best = fmin(objective, space, algo=tpe.suggest, max_evals=max_evals)
    best_params = space_eval(space, best)
    return best_params

def optimise_models(space_dict, max_evals=100):
    best_params_dict = {}
    for model_name, space in space_dict.items():
        best_params = optimise_model(model_name, space, max_evals=max_evals)
        best_params_dict[model_name] = best_params
    return best_params_dict

best_params_dict = optimise_models(space_dict, max_evals=100)
```

### With cross-validation

```python
from sklearn.metrics import average_precision_score
from hyperopt import fmin, tpe, STATUS_OK, space_eval
import pickle
reload(h)

def optimise_model(model_name, space, max_evals=100):
    # Map the model name to the corresponding classifier
    model_dict = {
        'xgb': XGBClassifier,
        'glm': LogisticRegression,
        'rf': RandomForestClassifier,
        'svm': SVC
    }
    model = model_dict[model_name]

    def objective(params):
        # Load the appropriate dataset
        window_size = params.pop('window_size')
        dataset_name = h.get_dataset_name(window_size, reference_class)
        _, _, _, _, _, _, X_train_full, y_train_full, train_full_idx, _, _, _, _, _, participants = h.prepare_datasets(filename=f"output/{dataset_name}", test_size=0.1, val_size=0.1, verbose=False)

        # Create folds
        participants = participants[train_full_idx] # Subset participants
        folds = h.create_folds(X_train_val, y_train_val, groups=participants, n_folds=5) # Exhaustive would be 12 folds

        # Train and evaluate the model using cross-validation
        clf = model(**params, random_state=42)
        scores = []

        for train_index, test_index in folds:
            X_train_fold, X_test_fold = X_train_val[train_index], X_train_val[test_index]
            y_train_fold, y_test_fold = y_train_val[train_index], y_train_val[test_index]
            clf.fit(X_train_fold, y_train_fold)
            y_pred = clf.predict(X_test_fold)
            score = average_precision_score(y_test_fold, y_pred)
            scores.append(score)
        score = np.mean(scores)  # Use the average score

        return {'loss': -score, 'status': STATUS_OK, 'params': params}

    # Perform the optimisation
    best = fmin(objective, space, algo=tpe.suggest, max_evals=max_evals)
    best_params = space_eval(space, best)
    return best_params

def optimise_models(space_dict, max_evals=100):
    best_params_dict = {}
    for model_name, space in space_dict.items():
        best_params = optimise_model(model_name, space, max_evals=max_evals)
        best_params_dict[model_name] = best_params
    return best_params_dict

best_params_dict = optimise_models(space_dict, max_evals=100)

# Save the best parameters
with open('output/ml_best_params_cv_nt.pkl', 'wb') as f:
    pickle.dump(best_params_dict, f)
```

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler # It is very significant to remember for SVM that the input data need to be normalized so that features are on the same scale and compatible.

model_name = 'svm'

# Define the hyperparameter space
space = hp.choice('model_type', [
    {
        'window_size': hp.choice('window_size_linear', range(8, 34, 2)),
        'C': hp.loguniform('C_linear', np.log(0.1), np.log(100)),
        'kernel': 'linear',
        'class_weight' : 'balanced'
    },
    {
        'window_size': hp.choice('window_size_rbf', range(8, 34, 2)),
        'C': hp.loguniform('C_rbf', np.log(0.1), np.log(100)),
        'kernel': 'rbf',
        'gamma': hp.loguniform('gamma_rbf', np.log(0.0001), np.log(10)),
        'class_weight' : 'balanced'
    }
])

# Define the objective function to minimize
def objective(params):
    # Load the appropriate dataset
    window_size = params.pop('window_size')
    # params.pop('reference_class')
    dataset_name = h.get_dataset_name(window_size, reference_class)
    X_train, y_train, X_val, y_val, _, _, _, _, _, _, _, _, _, _, _ = h.prepare_datasets(filename=f"output/{dataset_name}", test_size=0.1, val_size=0.1, verbose=False)

    # Normalize the data
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    
    # Train and evaluate the model
    svm_model = SVC(**params, random_state=42)
    svm_model.fit(X_train, y_train)
    y_pred = svm_model.predict(X_val)
    score = average_precision_score(y_val, y_pred)

    # Add the window size and type back into params for logging
    params['window_size'] = window_size
    # params['reference_class'] = reference_class

    return {'loss': -score, 'status': STATUS_OK, 'params': params}

# Perform the optimization
best = fmin(objective, space, algo=tpe.suggest, max_evals=100)
best_params_svm = space_eval(space, best)
print("Best set of hyperparameters: ", best_params_svm)

# Save the best hyperparameters
if reference_class == 'nt':
    ml_best_params_nt[model_name] = best_params_svm
else:
    ml_best_params_tnt[model_name] = best_params_svm
```
